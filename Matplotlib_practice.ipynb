```python
import matplotlib.pyplot as plt
```

## ***Simple Example of Making Graphs***

```python
days = [1,2,3,4,5,6,7]
likes = [190,210,220,150,200,300,70]
views = [300,400,600,200,400,700,100]
```

```python
plt.title("Posts per likes:")
plt.xlabel("No. of Days") #Giving label to x-axis
plt.ylabel("No. of Likes per day") #Giving label to y-axis
plt.plot(days,likes,marker='s')
plt.grid(color='green') #It forms grid on the chart also we can change the color of grid too
```

```python
plt.title("Posts per views per day")
plt.xlabel("No. of Days")
plt.ylabel("No. of Views per day")
plt.plot(days,views,color='red',marker='o',mfc='black',mec='yellow',ls='--',lw=3,ms=8)
#There are different ways to represt the  graph line by using 'ls'.
#And to change marker i.e. pointer we use 'marker' and its properties can be changed by using 'mfc','mec','ms',etc.
#ls=line size, lw=line width, mfc=marker faceColor, mec=marker edgeColor, ms=marker size, etc
```

```python
#If we want limited graph i.e. graph of limited data then:
plt.plot(days,views,)
plt.xlim(3,5)
plt.ylim(300,700)
```

## ***Comparing different social media platform data.***

```python
days = [1,2,3,4,5,6,7]
fb_likes = [190,210,220,350,200,300,70]
insta_likes = [710,1250,565,875,420,735,550]
youtube_likes = [1000,350,855,1700,800,550,950]
linkedin_likes = [10,50,65,170,20,35,50]
```

```python
plt.title("Social media Analyzation:")
plt.xlabel("No. of Days:")
plt.ylabel("No. of Likes:")
plt.plot(days,fb_likes,marker='o',label='Facebook',color='blue')
plt.plot(days,insta_likes,marker='s',label='Instagram',color='black')
plt.plot(days,youtube_likes,marker='d',label='Youtube',color='red')
plt.plot(days,linkedin_likes,marker='p',label='LinkedIn',color='m')
plt.legend()
plt.grid()
#Saving the graph at desired location:
plt.savefig("Graph.png")
```

```python
plt.plot(days,fb_likes,marker='o',label='Facebook',color='blue')
plt.plot(days,insta_likes,marker='s',label='Instagram',color='black')
plt.plot(days,youtube_likes,marker='d',label='Youtube',color='red')
plt.plot(days,linkedin_likes,marker='p',label='LinkedIn',color='m')
plt.legend()
plt.grid()
#Finding likes between 300-800 from day 1 to day 5
plt.ylim(300,800)
plt.xlim(1,5)
```

## ***Different types of graph that can be created:***

```python
#Scatter plot:
days = [1,2,3,4,5,6,7]
plots = likes.copy()
plt.xlabel("No. of days:")
plt.ylabel("plots:")
plt.scatter(days,plots,color="black",marker='s',s=50)
plt.grid()
plt.show()
```

```python
plt.xlabel("No. of days:")
plt.ylabel("plots:")
plt.scatter(days,plots,marker='s',s=50,cmap='Oranges',c=[1,2,3,4,5,6,7])
plt.grid()
#For making the markrs from light to dark of same color we use color-map i.e. cmap function
```

```python
#Bar Graph:
days = [1,2,3,4,5,6,7]
plots = likes.copy()
plt.bar(days,plots,color=['red','yellow','blue','green','orange','black','m'])
```

```python
#Pie Chart:
likes=[250,190,470,550,390]
platform=['Instagram','LinkedIn','Facebook','Youtube','Twitter']
plt.pie(likes,colors=['m','blue','cyan','red','orange'],labels=platform,autopct="%1.1f",explode=[0.2,0,0,0,0],shadow=True)
plt.show()
```
